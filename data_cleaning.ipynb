```python
import json
import re
from bs4 import BeautifulSoup

# Load the scraped JSON data
with open('scraped_data.json') as f:
    scraped_json = json.load(f)

def clean_text(summary):
    """Function to clean and normalize text."""

    summary = BeautifulSoup(summary, "html.parser").get_text()

    summary = re.sub(r"\s+", " ", summary)

    summary = re.sub(r"\\t|\\n|\\u200b", " ", summary)

    summary = re.sub(r"[^\w\s.,!?;:'\"()-]", "", summary)

    summary = summary.lower()

    summary = summary.strip()
    
    return summary

def build_text_files(data_json, dest_path):
    """Function to process JSON data and save it to a text file."""
    with open(dest_path, 'w') as f:
        data = ''
        for texts in data_json:
            if texts and 'content' in texts and 'p' in texts['content']:
                summary = str(texts['content']['p']).strip()
                cleaned_summary = clean_text(summary)
                data += cleaned_summary + "  "

        print(f'{dest_path} length: {len(data)} characters')
        f.write(data)

build_text_files(scraped_json, 'cleaned_dataset.txt')
```
